### Czarno biały filtr

```python
# na podstawie jednej z warstw przepisac klolory do innych
im2= im1[:,:,1]=im1[:,:,0] 
im2 = im1[:,:,2] = im1[:,:,0]   
plt.imshow(im2)

#inna metoda
suma_z_warstw = im1.sum(axis=2, keepdims=True)
suma_z_warstw = suma_z_warstw /4    #np. max(suma z warstw)
data = np.ones(3) * suma_z_warstw
data.shape
plt.imshow(data)
```
